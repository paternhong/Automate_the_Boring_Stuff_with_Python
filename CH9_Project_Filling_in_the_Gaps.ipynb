Write a program that finds all files with a given prefix, such as spam001.txt, spam002.txt, and so on, in a single folder and locates any gaps in the numbering (such as if there is a spam001.txt and spam003.txt but no spam002.txt). Have the program rename all the later files to close this gap.

As an added challenge, write another program that can insert gaps into numbered files so that a new file can be added

```python
#! /usr/bin/env python3
import os, re, shutil

while True:
    folder = input('Please input the folder to change file name: \n')
    if os.path.isdir(folder):
        break
    print('Please enter the correct path.')
os.chdir(folder)
# os.chdir('/Users/paternhong/Desktop/ML100/1_Automate_the_Boring_Stuff_with_Python/CH9_Project_Filling_in_the Gaps')
name = input('Input file name: \n')
fileType = input('input file type (ex. .txt, .png ....): \n')
numbering = int(input("Input the numbers of the numbering (ex. for 001235, it will be 6)"))
order = input("Input numbering position (Pre = 1, Post = 2): \n")
while order not in ['1', '2']:
    print('Please select correct numbering position.')
    order = input("Input numbering position (Pre = 1, Post = 2): \n")
# 列出 folder 內 files
filenames = os.listdir()

if order == '1':
    combine = '^' + '\\d{' + str(numbering) + '}' + name + fileType + '$'
    find = re.compile(combine)

    # 創造符合規則的待確認檔案 list，接著找出 folder 內符合規則的檔案並新增到 newList
    newList = []
    for file in filenames:
        if find.search(file):
            newList.append(file)
    print('{} files correspond the request.'.format(len(newList)))
    
    # 開始確認編號並修正，從 001 開始直到符合規則的檔案數目！
    for num in range(1, len(newList) + 1):
        file_now = str(num).zfill(numbering) + name + fileType
        # 先判斷此編號是否存在
        if file_now in newList:
            print('The file {} has numbered correctly.'.format(file_now))
            # 此編號 file 已存在 newList 內，從待確認 list 移除此編號 file，並直接進行下一個編號
            newList.remove(file_now)
            continue
        # 此編號不存在，找最接近的下一個，loop到找到為止
        next_num = num + 1
        while str(next_num).zfill(numbering) + name + fileType not in newList:
            next_num += 1
        # 把最接近的檔案命名成 file_next
        file_next = str(next_num).zfill(numbering) + name + fileType

        # 此時 next_num 應該為最接近的下一個編號，檔案為 file_next
        # 檔案 file_next 更名為 file_now
        print('Changing the the name of {} to {}'.format(file_next, file_now))
        shutil.move(file_next, file_now)
        # 從待確認 list 移除 file_next 字串，因為此檔案已被確認過並正確變更編號為當前號碼
        newList.remove(file_next)

if order == '2':
    combine = '^' + name + '\\d{' + str(numbering) + '}' + fileType + '$'
    find = re.compile(combine)

    # 創造符合規則的待確認檔案 list，接著找出 folder 內符合規則的檔案並新增到 newList
    newList = []
    for file in filenames:
        if find.search(file):
            newList.append(file)
    print('{} files correspond the request.'.format(len(newList)))
    
    # 開始確認編號並修正，從 001 開始直到符合規則的檔案數目！
    for num in range(1, len(newList) + 1):
        file_now = name + str(num).zfill(numbering) + fileType
        # 先判斷此編號是否存在
        if file_now in newList:
            print('The file {} has numbered correctly.'.format(file_now))
            # 此編號 file 已存在 newList 內，從待確認 list 移除此編號 file，並直接進行下一個編號
            newList.remove(file_now)
            continue
        # 此編號不存在，找最接近的下一個，loop到找到為止
        next_num = num + 1
        while name + str(next_num).zfill(numbering) + fileType not in newList:
            next_num += 1
        # 把最接近的檔案命名成 file_next
        file_next = name + str(next_num).zfill(numbering) + fileType

        # 此時 next_num 應該為最接近的下一個編號，檔案為 file_next
        # 檔案 file_next 更名為 file_now
        print('Changing the the name of {} to {}'.format(file_next, file_now))
        shutil.move(file_next, file_now)
        # 從待確認 list 移除 file_next 字串，因為此檔案已被確認過並正確變更編號為當前號碼
        newList.remove(file_next)
```

```python
#! /usr/bin/env python3
'''
把符合條件的檔案全部強制更名 or 副檔名，並"保留"原本編號！
'''
import os, re, shutil

while True:
    folder = input('Please input the folder to change file name: \n')
    if os.path.isdir(folder):
        break
    print('Please enter the correct path.')
os.chdir(folder)

while True:
    order = input("Input file order type (Pre-num = 1, Post-num = 2, Pre to Post = 3, Post to Pre = 4): \n")
    if order in ['1', '2', '3', '4']:
        break
    print('Please input the legal order type.')

name = input('Input file name: \n')
fileType = input('Input file type (ex. .txt, .png ....): \n')
newName = input('Input new file name: \n')
newFileType = input('input new file type (ex. .txt, .png ....): \n')

if order == '1':
    # combine 部分可以依照需求變更
    combine = '^(\\d*)(' + name + ')(' + fileType + ')$'  # 形式類似  '^(name)(\\d{3})(fileType)$' 利用括號分組
    find = re.compile(combine)
    filenames = os.listdir()

    newList = []
    # 找出符合規則的檔案清單
    for file in filenames:
        if find.search(file):
            newList.append(file)

    # 用符合規則的檔案清單來修改
    if len(newList) == 0:
        print("There's no correspond files in the folder.")
    else:
        for file in newList:
            partList = find.search(file)
            print('Changing file {} to {}'.format(file, newName + partList.group(1) + newFileType))
            shutil.move(file, newName  + partList.group(1) + newFileType)

if order == '2':
    # combine 部分可以依照需求變更
    combine = '^(' + name + ')(\\d*)(' + fileType + ')$'  # 形式類似  '^(name)(\\d{3})(fileType)$' 利用括號分組
    find = re.compile(combine)
    filenames = os.listdir()

    newList = []
    # 找出符合規則的檔案清單
    for file in filenames:
        if find.search(file):
            newList.append(file)

    # 用符合規則的檔案清單來修改
    if len(newList) == 0:
        print("There's no correspond files in the folder.")
    else:
        for file in newList:
            partList = find.search(file)
            print('Changing file {} to {}'.format(file, newName + partList.group(2) + newFileType))
            shutil.move(file, newName + partList.group(2) + newFileType)

if order == '3':
    # combine 部分可以依照需求變更
    combine = '^(\\d*)(' + name + ')(' + fileType + ')$'  # 形式類似  '^(name)(\\d{3})(fileType)$' 利用括號分組
    find = re.compile(combine)
    filenames = os.listdir()

    newList = []
    # 找出符合規則的檔案清單
    for file in filenames:
        if find.search(file):
            newList.append(file)

    # 用符合規則的檔案清單來修改
    if len(newList) == 0:
        print("There's no correspond files in the folder.")
    else:
        for file in newList:
            partList = find.search(file)
            print('Changing file {} to {}'.format(file, newName + partList.group(1) + newFileType))
            shutil.move(file, newName  + partList.group(1) + newFileType)            

if order == '4':
    # combine 部分可以依照需求變更
    combine = '^(' + name + ')(\\d*)(' + fileType + ')$'  # 形式類似  '^(name)(\\d{3})(fileType)$' 利用括號分組
    find = re.compile(combine)
    filenames = os.listdir()

    newList = []
    # 找出符合規則的檔案清單
    for file in filenames:
        if find.search(file):
            newList.append(file)

    # 用符合規則的檔案清單來修改
    if len(newList) == 0:
        print("There's no correspond files in the folder.")
    else:
        for file in newList:
            partList = find.search(file)
            print('Changing file {} to {}'.format(file,  partList.group(2) + newName + newFileType))
            shutil.move(file, partList.group(2) + newName + newFileType)
```
