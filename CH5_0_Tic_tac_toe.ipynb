```python
class tic_tac_toe():
    def __init__(self):
        self.board = {'top_l': ' ', 'top_m': ' ', 'top_r': ' ',
        'middle_l': ' ', 'middle_m': ' ', 'middle_r': ' ',
        'bottom_l': ' ', 'bottom_m': ' ', 'bottom_r': ' '}
        
    def drawTheBoard(self, board):
        print(board['top_l'] + '|' + board['top_m'] + '|' + board['top_r'])
        print('-+-+-')
        print(board['middle_l'] + '|' + board['middle_m'] + '|' + board['middle_r'])
        print('-+-+-')
        print(board['bottom_l'] + '|' + board['bottom_m'] + '|' + board['bottom_r'])
        
    def whoWin(self, board):
        winner = None
        board_table = list(board.values())
        for i in range(3):
            if board_table[3 * i : 3 * i + 3].count('O') >= 1 and board_table[3 * i : 3 * i + 3].count('X') >= 1:
                pass
            elif board_table[3 * i : 3 * i + 3].count('O') == 3:
                winner = 'O'
                return winner
            elif board_table[3 * i : 3 * i + 3].count('X') == 3:
                winner = 'X'
                return winner
        for i in range(3):
            if board_table[i: i + 7: 3].count('O') >= 1 and board_table[i: i + 7: 3].count('O') >= 1:
                pass
            elif board_table[i: i + 7: 3].count('O') == 3:
                winner = 'O'
                return winner
            elif board_table[i: i + 7: 3].count('X') == 3:
                winner = 'X'
                return winner
        if board_table[0: 9: 4].count('O') == 3 or board_table[2: 7: 2].count('O') == 3:
            winner = 'O'   
        elif board_table[0: 9: 4].count('X') == 3 or board_table[2: 7: 2].count('X') == 3:
            winner = 'X'
        return winner
    
    def playGame(self):
        win = None
        turn = 'X'
        self.drawTheBoard(self.board)
        for i in range(9):
            in_pos = input('It\'s ' + turn + ' turn: ')
            while in_pos not in self.board.keys():
                print('Please position is incorrect!')
                in_pos = input('please enter correct position: ')
            while self.board[in_pos] != ' ':
                print('The cell had been filled')
                in_pos = input('please choose another position: ')
            self.board[in_pos] = turn
            self.drawTheBoard(self.board)
            win = self.whoWin(self.board)
            if win:
                break
            if turn == 'X':
                turn = 'O'
            elif turn == 'O':
                turn = 'X'
        if win:
            print('The ' + turn + ' side wins the game')
        else:
            print('The game ended in a draw')
    

    
```

```python
def play():
    tic_tac_toe().playGame()
play()
```
